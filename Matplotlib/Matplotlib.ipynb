```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,25)
y=np.arange(10,35)

plt.scatter(x,y,c="b",marker=".")

plt.title("Scatter plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
plt.plot(x,y,c="red")
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
x=np.arange(0,np.pi,0.1)
y=np.sin(x)
print("values from 0 to 3.14 =",x)
print("")
print("sin values from 0 to 3.14 =",y)
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)

plt.plot(x,y,"black")
plt.title("Sin wave")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.cos(x)

plt.plot(x,y,"blue")
plt.title("Cos wave")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.tan(x)

plt.plot(x,y,"red")
plt.title("Tan wave")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
months=np.array(["jan","feb","march","april","may","june","july"])
No_of_Refrigerators_Sold=np.array([26,65,48,29,65,63,55])

plt.bar(months,No_of_Refrigerators_Sold)
plt.title("Months vs sales")
plt.xlabel("Months")
plt.ylabel("Sold refrigerators")
plt.show()
```

```python
x = np.array([67,44,34,7,90,8,34,76,23,85,12,72,3,64,38,5]) 
plt.hist(x)
plt.show()
```

```python
x = np.array([67,44,34,7,90,8,34,76,23,85,12,72,3,64,38,5]) 
plt.hist(x,bins=10)
plt.xlabel('X values')
plt.ylabel('No. of values')
plt.xticks(np.arange(0,100,5))
plt.yticks(np.arange(0,5))
plt.show()
```

```python
data=pd.Series([10,20,54,89,62,14,36,65,200,-3])
plt.boxplot(data)
plt.show()
```

```python
data=[65,32,85,20]
label=['cabbage','tomato','potato','corriander']
colors = ['r', 'g', 'b', 'y']
explode=(0,0,0.1,0.23)

plt.pie(data,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('Vegetables vs sales')
plt.show()
```

```python
x1=np.arange(10)
y1=x1**2

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x1,y1,'bo')
plt.title("fig 1")

plt.subplot(2,2,2)
plt.plot(x1,y1,'go')
plt.title("fig 2")

plt.subplot(2,2,3)
plt.plot(x1,y1,'D')
plt.title("fig 3")

plt.subplot(2,2,4)
plt.plot(x1,y1,'g*--',)
plt.title("fig 4")
plt.show()
```
